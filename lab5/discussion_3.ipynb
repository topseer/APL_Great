```python
% matplotlib inline
import pandas as pd
import numpy as np
import random as py_random
import numpy.random as np_random
import time
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

sns.set(style="whitegrid")
```

```python
insurance = pd.read_csv("insurance.csv")
```

```python
#convert the categortical variable from object type to int

insurance["sex"] = insurance["sex"].replace(["female","male"], [0,1])
insurance["smoker"] = insurance["smoker"].replace(["no","yes"], [0,1])
insurance["region"] = insurance["region"].replace(["southwest","southeast","northwest","northeast"], [0,1,2,3])
```

```python
insurance["children"].describe()
```

According to the descriptive statistic, people with insurance has an average of 1 child. 50% has less than 1 child, and 75% has less than 2 children. This indicates that people with lot of children 3 or more don't really have insurance

```python
insurance["children"].value_counts().sort_index()
```

The counts support the argument above 

```python
counts = insurance["children"].value_counts().sort_index()
x = range( len( counts))
width = 1/1.5
figure = plt.figure(figsize=(20, 6))
axes = figure.add_subplot(1, 1, 1)
axes.bar(x, counts, width, color="dimgray", align="center")
axes.set_xticks(x)
axes.set_xticklabels(counts.axes[0])
axes.set_title( "counts vs. number of children")
axes.set_xlabel("children")
axes.set_ylabel( "Counts")
axes.xaxis.grid(False)
plt.show()
plt.close()
```

This clearly indicate the trend of people with more children vs number of insurance. The more children people have, the less likely they will buy insurance

```python
counts = insurance["children"].value_counts().sort_index()
proportions = insurance["children"].value_counts(normalize=True).sort_index()
x = range( len( counts))
width = 1/1.5
figure = plt.figure(figsize=(10, 6))
axes = figure.add_subplot(1, 1, 1)
axes.bar(x, proportions, width, color="dimgray", align="center")
axes.set_xlabel("Children")
axes.set_xticks(x)
axes.set_xticklabels(proportions.axes[0])
axes.set_title( "Relative Frequency of children")
axes.set_ylabel( "Percent")
axes.xaxis.grid(False)
plt.show()
plt.close()
```

This is similar to the above bar graph but in percentage unit of the overall population
