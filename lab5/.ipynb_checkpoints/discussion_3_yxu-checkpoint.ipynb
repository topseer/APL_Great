```python
import numpy as np
import random as py_random
import numpy.random as np_random
import time
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
sns.set(style="whitegrid")

insurance = pd.read_csv("insurance.csv")
insurance.head()
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

#distribution of insurance charges
plt.figure()
ax = sns.distplot(insurance.charges); 

#distribution of insurance bmi
plt.figure()
ax = sns.distplot(insurance.bmi); 

#distribution of insurance age
plt.figure()
ax = sns.distplot(insurance.age); 
```

```python
insurance.groupby(['age']).mean()["charges"]

#scatter of charges by age, smoker vs non-smoker
ax =  sns.scatterplot(x=insurance.age, y=insurance.charges,hue=insurance.smoker)
```

```python
insurance_Smoker = insurance.loc[insurance.smoker == "yes"]
insurance_noSmoker = insurance.loc[insurance.smoker == "no"]


#Smokers only:   scatter of charges by age, color: bmi
plt.figure()
ax = sns.scatterplot(x=insurance_Smoker.age, y=insurance_Smoker.charges, hue=insurance_Smoker.bmi)


#Non-Smokers only:   scatter of charges by age, color: bmi
plt.figure()
ax = sns.scatterplot(x=insurance_noSmoker.age, y=insurance_noSmoker.charges, hue=insurance_noSmoker.bmi)
```
