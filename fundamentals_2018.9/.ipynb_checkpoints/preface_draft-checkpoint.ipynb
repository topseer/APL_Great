# Preface

There are the notes for a semester long course in Data Science.
Unlike the typical graduate course in programming or artificial intelligence, the syllabus for a data science course is far from settled.

In 2013, I taught the first class in Data Science at Johns Hopkins University's Engineering for Professional's Computer Science degree program.
The course has since moved to its own degree, a Master's degree in Data Science.
Five years later, there still isn't a lot of clarity about what Data Science is or should be.

These are the class notes for a semester long course in data science.

At the present time there is no *single* textbook that addresses the fundamental concepts and techniques of Data Science suitable for a graduate level course in the subject.
In fact, there aren't really any textbooks about Data Science at all.
Most of the books available are published by [O'Reilly](https://www.oreilly.com/) or [Manning](https://www.manning.com/) and can be uneven in their quality or relevance.
In some cases, in order to decide whether a particular chapter is "good", you often have to already know what the chapter is talking about.
In other cases, the chapter emphasizes a tool rather than a technique or concept that is tool independent.

So while these are often very good books (I learn new things from them all the time), they aren't really suitable as textbooks.
After all, graduate school is not vocational training and we should not be concerned with the "Flavor of the Month" (Tableau or Superset? Hadoop or Spark?) but instead the concepts that transcend tools.
And while I have the background to understand when a topic is presented incorrectly (most of these authors are self-taught), I have no desire to fight against such a text for 14 weeks.

We could take a different tack.
Data Science is a multi-disciplinary field and there are certainly graduate level textbooks on visualization, machine learning, statistics, databases, and so on.
Here we run into a different kind of problem.
How do we pick and chose from 6-10 textbooks each of which is designed for one or more semesters and reduce them to their essence for a single class?
How do we justify students paying for 6-10 textbooks?
I can't; I couldn't.

As a result, when the Data Science course was started in 2013, I taught from course notes in the form of Jupyter Notebooks.
This gave me the freedom to teach the topics as a I saw fit and to reduce those topics to their fundamentals.
This has met with mixed success.
Students prefer books; it's not easy to have constant access to the Jupyter Notebooks.
You can't read them on an iPhone on a subway.
As a result, I started compiling those course notes into a book form while keeping them as Jupyter Notebooks.

When trying to teach and learn Data Science, we run into a second and perhaps more substantial problem.
I'm not entirely sure we know what Data Science is or what it will be in 10 years.
There are those that say "Data Science" is just a sexy name for statistics.
I agree about 95%.

Hal Varian has a famous quote
about the sexiness of statistics in the coming decade:

> I keep saying the sexy
job in the next ten years will be statisticians. People think I’m joking, but
who would’ve guessed that computer engineers would’ve been the sexy job of the
1990s?

And he goes on to describe what are pretty much the daily activities of
a Data Scientist:

> The ability to take data – to be able to understand it, to
process it, to extract value from it, to visualize it, to communicate[. I]t’s
going to be a hugely important skill in the next decades, not only at the
professional level but even at the educational level for elementary school kids,
for high school kids, for college kids. Because now we really do have
essentially free and ubiquitous data. So the complimentary scarce factor is the
ability to understand that data and extract value from it.

>I think
statisticians are part of it, but it’s just a part. You also want to be able to
visualize the data, communicate the data, and utilize it effectively. But I do
think those skills – of being able to access, understand, and communicate the
insights you get from data analysis – are going to be extremely important.
Managers need to be able to access and understand the data themselves. [Flowing
Data](https://flowingdata.com/2009/02/25/googles-chief-economist-hal-varian-on-
statistics-and-data/)

But statistics has been a part of business for decades.
Gosset used statistics to improve the brewing process at Guinness in the early
1900's. Why "data science" and why now? I think there are several answers to
this. First, I think many of the original data scientists were just plain
ignorant of the existence of business statistics. Second, I think the deluge of
data was a watershed event in the broad applicability of business statistics.
Third, in many cases, academic statistics failed to provide what people wanted
from statistics. On this last point, Allen Downey is on target,

> Having read
"The Theory That Would Not Die" and "The Lady Tasting Tea" recently, I suggest
the following conjecture:

> The term "data scientist" has been created to
describe what people want from a statistician, but which many statisticians fail
to provide because statistics, as a field, spent too much time in the 20th
century on problems in philosophy of science, and theoretical mathematical
problems, and not enough time on practical applications and the use of
computation to analyze data. As a result, many graduates from statistics
programs have a set of skills that is not a good match for what the market wants
them to do. This market vacuum is being filled by "data science."

> That's my
theory, which is mine. [Allen
Downey](https://www.reddit.com/r/statistics/comments/1jygwl/is_data_scientist_just_a_sexed_up_word_for/)

This falls into the "Two Cultures" theory of machine learning and statistics.
Simon Blomberg said,

> From R's fortunes package: To paraphrase provocatively,
'machine learning is statistics minus any checking of models and assumptions'.
-- Brian D. Ripley (about the difference between machine learning and
statistics) useR! 2004, Vienna (May 2004) :-) Season's Greetings! [Two
Cultures](https://stats.stackexchange.com/questions/6/the-two-cultures-
statistics-vs-machine-learning)

To which Andrew Gelman responded,

> In that
case, maybe we should get rid of checking of models and assumptions more often.
Then maybe we'd be able to solve some of the problems that the machine learning
people can solve but we can't!

But this explanation seems incomplete. If you
look at actual Data Scientist job descriptions, they often include a large chunk
of good old fashioned software engineering and a greater focus on machine
learning.

This tension is echoed in a tweet by Josh Wills:

> Data Scientist
(n.): Person who is better at statistics than any software engineer and better
at software engineering than any statistician.
[@josh_wills](https://twitter.com/josh_wills/status/198093512149958656?lang=en)
And then there's a awful lot of hype surrounding what "Data Scientists" can do
for the organization--a hype that perhaps is rivaled only by the machinations of
the [Underwear Gnomes](https://en.wikipedia.org/wiki/Gnomes_(South_Park):

1. Hire data scientist.
2. ???
3. Profit

Let's take the following job posting as
an example:

> [Company] is looking for a Data Scientist eager to use advanced
analytical, machine learning and data transformation techniques as a means to
develop practical tools and analyses that can help solve complex business
problems; transforming volumes of data into actionable information. In some
instances, you will be using languages like R or Python to employ newer
techniques like tree ensembles, neural nets, and clustering to solve long-
standing questions. **In other instances, it will be your responsibility to come
up with solutions for problems that have not yet been identified.** In order to
do so, the Data Scientist will have to be confident that they can solve a wide-
range of problems using a variety of techniques—some known, some new, some yet
to be created.

My favorite bit is the part I've emphasized in boldface:
basically, you must solve unidentified problems.
How is that even possible?

And yet another plot twist comes from the aforementioned O'Reilly, purveyors of
books and hype, in the form of the clickbait headline, "What's a machine
learning engineer? Move over, data scientist. [There's a new job title in
town.](https://www.oreilly.com/ideas/what-are-machine-learning-
engineers?imm_mid=0f28f9&cmp=em-data-na-na-newsltr_ai_20170612)"
This is
probably important for two reasons.
First, data science is not (only) machine
learning.
Machine learning is but one tool and not necessarily the most
important tool embedded in a process that emphasizes communication and solving
problems of a particular kind.
Second, machine learning is not data science.
Machine learning courses rarely teach any or most of the skills needed to be an
effective data scientist.

Hand wringing aside, what seems clear, however, is
that data *science* is not like biology or chemistry.
Biologists are scientists
whose domains are, of course, biology; and likewise for chemist, anthropologist,
sociologist, etc.
In contrast, I have:

1. analyzed patterns of compensation and work rules for police, fire fighters and correctional officers.
2. explored hiring patterns for racial discrimination in a construction trades hiring hall.
3. investigated relationships between grants reviewed by peer boards in medical research.
4. built a model for mapping from OSHA exposure report job titles to SOC job classifications.
5. engaged in topic modeling in medical grant proposals.
6. created personalization models for customers of daily deals with an ever rotating inventory.
7. engineered infrastructure for the development and deployment of personalization models.
8. constructed a metrics dashboard and analysis for a project management SasS company.
9. developed statistical models for project completion including epic and story completion estimation.

So while the domain knowledge, programming languages and tools changed from job to job, and project to project, every task has shared fundamentals: communication, statistics, programming, machine learning, etc.

At the end of the day, a course must be taught and it must have some level of consistency.
The view I have taken is that data science is *largely* concerned with the same problems of everyday life as business statistics but to that I have substituted the Bayesian approach to statistics in lieu of the Frequentist approach and, where applicable, added the tools of machine learning.

I think that, "practical applications and the use of computation to analyze data" *and communicating the results* sums it up nicely.
