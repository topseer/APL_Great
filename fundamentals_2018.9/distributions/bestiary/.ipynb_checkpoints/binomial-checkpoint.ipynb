# Binomial Distribution

There is perhaps no better distribution then Binomial distribution to drive home the point that probability is simply counting.

The basis of the Binomial distribution--the generating story--is Bernoulli trials, specifically, with probability $p$ of success, what is the probability of observing $k$ successes in $n$ trials or attempts. These trials can be simultaneous (you have a pool of 200 smokers, if the probability of quitting (success) is 20%, what's the probability of 100 of them quitting smoking?) or they can be sequential (what's the probability of getting 10 heads in 20 tosses of a fair coin)--as long as each trial is independent.

Generally speaking, we simply add up all the ways we can get the number of indicated successes divide by all possible outcomes ("normalize"). The calculation of the probability includes the (perhaps) famous, Binomial coefficient:

$PMF(x) = \binom{n}{k} p^k(1-p)^{n-k}$

The Method of Moments estimator for $p$ is:

$\hat{p} = m_1$

The parameters of the Binomial Distribution are:

$mean = np$

$variance = np(1-p)$

## Examples of the Binomial Distribution

Which leads us to an interesting question, what does the Binomial distribution look like? 

We can actually express it in terms of three parameters, more or less: $p$, $k$ and $n$. That is, we can plot, what is the probability of $k$ successes in $n$ trials as $p$ changes, what is the probability of $k$ successes given $p$ as $n$ changes ($n \geq p$), or what is the probability of $k$ successes given $p$ as $n$ changes. In general, we can fix two parameters and vary a third.

Below we plot three different Binomial distributions changing only $p$ between plots. $n$ is fixed at 20 and $k$, the number of successes, is the "free" variable (x-axis).

```python
%matplotlib inline
```

```python
from scipy.stats import binom
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
figure = plt.figure(figsize=(20,6))

n = 20
p = 0.25
ks = [binom.pmf( k, n, p) for k in range( 0, n + 1)]
axes = figure.add_subplot(1, 3, 1)
axes.bar(range(len( ks)), ks, alpha=0.75)
axes.set_title( "Binomial Distribution $p=%s$" % p)
axes.set_ylabel( "Probability Mass")
axes.set_xlabel( "Successes, $k$, in $n=%s$ obs." % n)

n = 20
p = 0.5
ks = [binom.pmf( k, n, p) for k in range( 0, n + 1)]
axes = figure.add_subplot(1, 3, 2)
axes.bar(range(len( ks)), ks, alpha=0.75)
axes.set_title( "Binomial Distribution $p=%s$" % p)
axes.set_ylabel( "Probability Mass")
axes.set_xlabel( "Successes, $k$, in $n=%s$ obs." % n)

n = 20
p = 0.75
ks = [binom.pmf( k, n, p) for k in range( 0, n + 1)]
axes = figure.add_subplot(1, 3, 3)
axes.bar(range(len( ks)), ks, alpha=0.75)
axes.set_title( "Binomial Distribution $p=%s$" % p)
axes.set_ylabel( "Probability Mass")
axes.set_xlabel( "Successes, $k$, in $n=%s$ obs." % n)
```

## Derivation from the Exponential Distribution

Although we've already described the generating story for the Binomial distribution above. I want to continue with a derivation that builds on the previous one for the Exponential distribution.

For the example, we consider tasks that need to get done. These tasks are completed at an average rate, $\beta$, of 2 per day.

```python
from scipy.stats import expon
```

```python
beta = 2.0

figure = plt.figure(figsize=(10,6))

axes = figure.add_subplot(1, 1, 1)
axes.set_title( "Exponential Distribution with $\lambda=%.4f$" % (1.0/beta))
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "Days to Complete Task")
axes.set_xlim(0, 10)
xs = [expon.pdf( k, 0, beta) for k in range(0, 11)]
axes.plot( xs, color="steelblue")
```

We consider all tasks completed in 1 day or under to be "excellent" and the remainder to be "average":

```python
beta = 2.0

figure = plt.figure(figsize=(10,6))

axes = figure.add_subplot(1, 1, 1)
axes.set_title( "Exponential Distribution with $\lambda=%.4f$" % (1.0/beta))
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "Days to Complete Task")
axes.set_xlim(0, 10)
ys = [expon.pdf( k, 0, beta) for k in range(0, 11)]
axes.plot( ys, color="steelblue")
xs = list(range(0, 11))
indicator = [x < 2 for x in xs]
axes.fill_between( xs, ys, where=indicator, color="steelblue", alpha=0.5)
```

This sets up two zones of the Exponential distribution for success and failure. What is the probability of success?

```python
expon.cdf( 1, 0, beta)
```

Remember the *cumulative* density function of any probability distribution gives the probability of a value equal to x *or smaller*. So CDF(1) here is the probability of a task taking 1 day or less given an average rate of 2 days to finish.

Let's say this process happens every Sprint (10 working days) and we have 20 tasks per sprint. Tasks are independent. Any task that takes more than 10 days is assumed to take 10 days. We want to see how 10 Sprints go. Basically, we need to simulate from the Exponential distribution, how long each of 20 Tasks takes. These are the $n=20$ trials.

We then take all Tasks at 1 day or under to be "success" and any other to be "not a success". We count these up.

To get a random sample from a particular Exponential distribution, we use `rvs`:

```python
expon.rvs( 0, beta, 1)
```

so this task took three-quarters of a day. Let's see what 20 tasks does:

```python
expon.rvs( 0, beta, 20)
```

Now we want to plot them on our Exponential distribution:

```python
figure = plt.figure(figsize=(10,6))

axes = figure.add_subplot(1, 1, 1)
axes.set_title( "Exponential Distribution with $\lambda=%.4f$" % (1.0/beta))
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "Days to Complete Task")
axes.set_xlim(0, 10)
ys = [expon.pdf( k, 0, beta) for k in range(0, 11)]
axes.plot( ys, color="steelblue")
xs = list(range(0, 11))
indicator = [x < 2 for x in xs]
axes.fill_between( xs, ys, where=indicator, color="steelblue", alpha=0.5)
tasks = expon.rvs( 0, beta, 20)
y_tasks = [expon.pdf( t, 0, beta) for t in tasks]
for t, y in zip(tasks, y_tasks):
    color = "steelblue" if t <= 1 else "firebrick"
    axes.plot( [t, t], [0, y], color=color)
```

We can now see the successes and failures against the generating Exponential distribution. Let's see what 10 of them look like:

```python
figure = plt.figure(figsize=(20,10))

for i in range( 10):
    axes = figure.add_subplot(2, 5, i+1)
    axes.set_title( "Exponential Distribution with $\lambda=%.4f$" % (1.0/beta))
    axes.set_ylabel( "Probability Density")
    axes.set_xlabel( "Days to Complete Task")
    axes.set_xlim(0, 10)
    ys = [expon.pdf( k, 0, beta) for k in range(0, 11)]
    axes.plot( ys, color="steelblue")
    xs = list(range(0, 11))
    indicator = [x < 2 for x in xs]
    axes.fill_between( xs, ys, where=indicator, color="steelblue", alpha=0.5)
    tasks = expon.rvs( 0, beta, 20)
    y_tasks = [expon.pdf( t, 0, beta) for t in tasks]
    for t, y in zip(tasks, y_tasks):
        color = "steelblue" if t <= 1 else "firebrick"
        axes.plot( [t, t], [0, y], color=color)
```

Now let's convert the days from the Exponential distribution into Successes and Failures for the Binomial distribution:

```python
np.sum([1 if x <= 1 else 0 for x in expon.rvs( 0, beta, 20)])
```

Finally, let's run a 10,000 Sprints, count the successes and plot them:

```python
figure = plt.figure(figsize=(10,6))

axes = figure.add_subplot( 1, 1, 1)
successes = [np.sum([1 if x <= 1.0 else 0 for x in expon.rvs( 0, beta, 20)]) for _ in range( 10000)]
bins = np.arange(0, 21, 0.5)
values, base = np.histogram( successes, bins=bins)
weighted_values, base = np.histogram( successes, bins=bins, weights=np.zeros_like(successes) + 1. / 10000)
axes.bar( base[:-1], weighted_values, color="steelblue")
axes.set_title( "Derived Binomial Distribution of Task Success")
axes.set_ylabel( "Probability Mass")
axes.set_xlabel( "Successful Tasks")
```

And now we see what the distribution is using the Binomial distribution directly:

```python
figure = plt.figure(figsize=(10,6))

n = 20
p = expon.cdf( 1, 0, beta)
ks = [binom.pmf( k, n, p) for k in range( 0, n + 1)]
axes = figure.add_subplot(1, 1, 1)
axes.bar(range(len( ks)), ks, color="steelblue")
axes.set_title( "Binomial Distribution $p=%s$" % p)
axes.set_ylabel( "Probability Mass")
axes.set_xlabel( "Successes, $k$, in $n=%s$ obs." % n)
```

which is nearly identical to the derived one.

One question you might be asking yourself is this. That's cool (I hope you're saying at least that), you've derived a Binomial distribution from an Exponential distribution for Task completion. But how do you derive the Exponential distribution from repeated observations of $n$ following Uniform distributions?

And that is, indeed, a good question.

My explanation is this. Tasks are generally composed of steps but a Task isn't complete according to the *minimum* of its steps completed. However, a Task is solvable by any number of $n$ solutions. If people generally try the shortest solution first then a Task is always solved in the shortest amount of time. This tends to make Tasks follow an Exponential distribution.

At least that's my story. Do you have a better one?

Empirically, I found that overall Tasks do generally follow an Exponential distribution; however, if you break them down in "difficulties" (or "Points" in Agile-speak) they tend to follow a Gamma distribution.

It's worth emphasizing that although I have derived the Binomial distribution from the Exponential and the Exponential from the Uniform, you don't have to do the same thing to use any of these distributions as models. The relationship to lower distributions may not be readily apparent for your particular process.
