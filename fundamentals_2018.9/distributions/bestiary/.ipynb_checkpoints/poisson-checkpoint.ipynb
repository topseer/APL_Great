# Poisson Distribution

The Poisson distribution is interesting because it deals with counts. One might reasonable ask, why is that interesting? The Binomial distribution deals with counts as well and you, my astute reader, would be correct. The Binomial distribution deals with $k$ successes in $n$ trials when the probability of success is $p$.

However, the Poisson distribution deals with successes *when the number of trials is unknown.* Put another way, it deals with counts when the number of exposures is unknown or unquantified. The Poisson distribution is named after someone, Sim√©on Denis Poisson, and he used it to model the number of fatal mule kicks in Prussian calvary each year. Once again, we are left with the strange situation where "success" is a kick in the head by a mule but ignoring that for now, we see we can't use the Binomial distribution because we have no idea how many "situations in which a soldier could get kicked in the head by a mule" existed in any given year.

The key assumption here is that the probability of each event, $p$, is very small so that the number of successes relative to the number of exposures is very small as well. And thus the Poisson distribution is good for describing probability distributions over counts of things in time or space.

* mule kicks in the Prussian calvary per year.
* grubs in a square meter of lawn.
* microbes in a liter of blood.
* mutations in a strand of DNA.
* goals in a soccer game.

You can think of each of these processes as combining very many Bernoulli trials or exposures, say, a grub in a square *centimeter* of lawn which individually have a low probability, $p$. As the number of trials, $n$, increases to "don't know" the number of successes, $k$, is still relatively small.

The Poisson distribution is parameterized by $\lambda$, the rate parameter, of events per interval (again, space or time). The notation for a random variable $X$ that is Poisson distributed is $X \sim Poisson(\lambda)$.

The Method of Moments estimate for $\lambda$ is:

$\hat{\lambda} = m_1$

For the Poisson distribution, we have the following:

$mean = \lambda$

$variance = \lambda$

Yes, for the Poisson distribution, the mean equals the variance.

As the Poisson distribution is just a special case of the Binomial distribution, the generating story emerges from the Exponential distribution where the rate is low and there are a lot of events. 

In this case, we simulate a (peculiar) case where there number of Tasks is large (over 500), the probability of finishing a Task in under 1 days is very low (each Task takes 100 days on average). This situation fits our needs for the Poisson distribution:

```python
%matplotlib inline
```

```python
from scipy.stats import expon, poisson
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
figure = plt.figure(figsize=(20,6))

beta = 50.0

axes = figure.add_subplot(1, 3, 1)
axes.set_title( "Exponential Distribution with $\lambda=%.4f$" % (1.0/beta))
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "Days to Complete Task")
axes.set_xlim(0, 100)
xs = list(range(0, 100))
indicator = [x < 2 for x in xs]
ys = [expon.pdf( k, 0, beta) for k in range(0, 100)]
axes.plot( ys, color="steelblue")
axes.fill_between( xs, ys, where=indicator, color="steelblue", alpha=0.5)

tasks = expon.rvs( 0, beta, 500)
y_tasks = [expon.pdf( t, 0, beta) for t in tasks]
for t, y in zip(tasks, y_tasks):
    color = "steelblue" if t <= 1 else "firebrick"
    axes.plot( [t, t], [0, y], color=color, alpha=0.5)

replications = [np.sum( [1 if t <= 1 else 0 for t in expon.rvs( 0, beta, 500)]) for _ in range( 1000)]
counts = np.bincount( replications)

axes = figure.add_subplot(1, 3, 2)
axes.set_title( "Derived Poisson Distribution from 1000 Replicates")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "Counts of Tasks completed in 1 Day or less")

axes.bar( list( range( len( counts))), counts)

axes = figure.add_subplot(1, 3, 3)
axes.set_title( "Exact Poisson Distribution with $\lambda=%.4f$" % np.mean( replications))
axes.set_ylabel( "Frequency")
axes.set_xlabel( "Counts of Tasks completed in 1 Day or less")
xs = list(range(0, 20))
ys = [poisson.pmf( x, np.mean( replications), 0) for x in xs]
axes.plot( xs, ys, color="steelblue", marker="o")
```

The first chart shows a single result from 500 Tasks with a mean completion rate of 50 days. We are calling "success" any Task that is completed in 1 day or less. As you might imagine, the probability of a "success" is low. The middle chart shows the results of 1000 replications of this experiment and represents the derived distribution of counts of "successes". This should be approximately Poisson distributed. The third chart shows the exact Poisson distribution using the Method of Moments estimate of $\lambda$.

Looking at a single set of trials, we can see that our criteria are fulfilled:

```python
print( "p( 1) = ", expon.cdf( 1, 0, beta))
completed = [1 if t <= 1 else 0 for t in tasks]
print( "mean:", np.mean( replications))
print( "variance:", np.var( replications))
```

The probability of one day or less is low (around 2%) and the mean and variance are nearly the same. The approximation improves as the number of replications increases.

## Extensions and Alternatives

Although you should always start with the simplest possible model, as you iterate, you may find that simple model is too simple. When using the Poisson distribution as a model, two possible problems are that there are more zero values than expected and violations of the equality of mean and variance.

### Zero-Inflated Poisson Distribution

It is sometimes the case that we are dealing with a process that generates more zero counts than the Poisson distribution would normally predict. In such cases, you can switch to a Zero-Inflated Poisson (ZIP) distribution. The ZIP distribution is interesting because it has two generating components. The first is a binomial generator which generates *structural* zeros (and is parameterized by $\pi$) and the second is a Poisson distribution that generates counts which may also include zeroes (and is parameterized by $\lambda$).

The Method of Moments estimators for the distribution are:

$\hat{\pi} = \frac{m_2 - m_1}{m_2 + m_1^2 - m_1}$

$\hat{\lambda} = \frac{m_2 + m_1^2}{m_1} - 1$

while the mean and variance are:

$mean = (1 - \pi)\lambda$

$variance = \lambda(1-\pi)(1+\pi\lambda)$

### Negative Binomial Distribution

There are many formulations for the Negative Binomial distribution and thus many generating stories. In the current context, we're interested in a specific parameterization which describes the waiting time in a Bernoulli process where there are $k$ successes, $r$ failures and a probability of success, $p$. In this case, the Negative Binomial distribution describes the probability of $k$ successes with the $r$-th failure on the *last* trial.

It turns out that this is a good model for a Poisson process where the variance exceeds the mean. Put differently, when the data are *overdispersed* relative to the Poisson distribution. Why might that happen? This happens in situations where the counts are "contagious" such as tornados during a storm, tropical cyclones during the hurricane season.

As in the Poisson distribution, we want to know the probability of a *success*, $k$, so $k$ is our random variable, $X$. Thus given our data, we are going to want to estimate $r$ -- how many failures there were and $p$ -- what the probability of success was.


$\hat{p} = \frac{m_1}{m_2}$

$\hat{r} = \frac{m_1^2}{m_2 - m_1}$

And the mean and variance are:

$mean = \frac{r(1-p)}{p}$

$variance = \frac{r(1-p)}{p^2}$
