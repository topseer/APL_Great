# Normal Distribution

The Normal or Gaussian distribution is probably the most famous of the Mathematical distributions. It is also known as the "bell shaped curve".

The basic generating story for the Normal distribution is that any process with small additive effects generally ends up with a Normal distribution. As with the Exponential distribution, this means we can generate the Normal distribution by combining the variables that are distributed according to the Uniform distribution. To simplify even further, we're going to have the same effect for all variables, $1.0$:

```python
import numpy as np
from numpy.random import uniform, seed
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
%matplotlib inline
```

```python
seed( 1738485857)
```

As we did before, we'll start out a single Uniform *variate* (a variable distributed according to the Uniform distribution). It doesn't really matter what it is, as long as it's small:

```python
xs1 = uniform(0, 1, 1000)

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1, 1, 1)
axes.hist(xs1, density=True, color="DimGray")
axes.set_title( "Uniform Distribution, $x_1$")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")

plt.show()
plt.close()
```

So that looks pretty Uniform.

We're going to add a few more variables in $x_2$ and $x_3$:

```python
xs2 = uniform(0, 1, 1000)
xs3 = uniform(0, 1, 1000)

xs = xs1 + xs2 + xs3
xs = xs - 1.5 # scale the average back to zero.

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1, 1, 1)
axes.hist(xs, density=True, color="DimGray")
axes.set_title( "Sum of Uniform Distributions, $x_1 + x_2 + x_3$")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")

plt.show()
plt.close()
```

Amazingly, with just three variables, we are already starting to see a Normal distribution emerge. What happens with 10 variables?

```python
xs = uniform( 0, 1, 1000)
for i in range( 9):
    xs = xs + uniform(0, 1, 1000)

xs = xs - 5 # scale the average back to zero.
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1, 1, 1)
axes.hist(xs, density=True, color="DimGray")
axes.set_title( "Sum of 10 Uniform Distributions")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")

plt.show()
plt.close()
```

A hundred variables?

```python
xs = uniform( 0, 1, 1000)
for i in range( 99):
    xs = xs + uniform(0, 1, 1000)

xs = xs - 50 # scale the average back to zero.
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1, 1, 1)
axes.hist(xs, density=True, color="DimGray")
axes.set_title( "Sum of 100 Uniform Distributions")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")

plt.show()
plt.close()
```

In truth, this distribution is called the [Irwin-Hall distribution](https://en.wikipedia.org/wiki/Irwin%E2%80%93Hall_distribution) but as the number of variables increases, it approaches the Uniform distribution. This is another example of the [Central Limit Theorem](https://en.wikipedia.org/wiki/Central_limit_theorem).

Later in the book we will see how a process that is the result of small increases and decreases is the foundation for Linear Regression.

The Normal distribution is parameterized by the mean (sometimes called $\mu$ or location) and the standard deviation ($\sigma$ or scale). And now we finally get to use our moments exactly as calculated:

$\hat{mu} = \bar{x} = m_1$

$\hat{\sigma} = s = \sqrt{m_2}$

And, of course, we have:

$mean = \mu = \frac{1}{n} \sum x_i$

$variance = \sigma^2 = \frac{1}{n} \sum (\mu - x_i)^2$

*standard deviation* $= \sigma = \sqrt{\sigma^2}$

The estimate of $\mu$ is so common, it has its own name $\bar{x}$ ("eks bar") instead of the more general "mu hat". Similarly, the estimate of the variance is called $s^2$ ("es squared") and standard deviation is just $s$ ("es").

Throughout this section we've seen different versions of the Central Limit Theorem. The most specific version says that sums of random variables tend towards a Normal distribution. We just saw this with sums of Uniform distributions and we saw it earlier with the Binomial distribution. The Poisson distribution with large $\lambda$ also tends towards the Normal distribution (the Poisson itself being a sum of Bernoulli trials).

The most general version of the Central Limit Theorem says that the resulting distribution depends on how you combine the random variables. If you take the minimum, as we did with the Uniform variables in the section of the Exponential distribution, you get the Exponential distribution.

Something similar happens when random variables combine in terms of proportional effects. You end up with the Log-Normal distribution.

## Log-Normal distribution

The Log-Normal distribution often useful because the domain of X is constrained to be *positive*. This is important for variables that cannot be negative such as height or purchases. However, this does not mean you always need to model non-negative variables with a Log-Normal distribution. Sometimes you can use a Normal distribution just fine as the mean and standard deviation are such that you're so far from 0 you're unlikely to predict someone has a negative height. As we will see throughout this book, "it depends".

The Log-Normal and Normal distributions are related to each other. If $X$ has a Log-Normal distribution then the natural log of $X$, $Y = ln(X)$ will have a Normal distribution. Conversely, if $Y$ has a Normal distribution, then $X = exp(Y)$ will have a Log-Normal distribution. And this sets up the generating story for the Log-Normal distribution: the Log-Normal distribution often results from processes where the random variable influence the outcome in a *multiplicative* or proporational way instead of an additive way. Height is a good example here. It's probably a good bet that different factors affect your height not in specific inches but in percentages of some base.

The Method of Moments estimators for the Log-Normal distribution are:

$\bar{x} = ln(\frac{m_1^2}{\sqrt{m_2 + m_1^2}})$

$s^2 = ln( 1 + \frac{m_2}{m_1^2})$

while the mean and variance are:

$mean = e^{\mu + \frac{\sigma^2}{2}}$

$variance = e^{2\mu + \sigma^2}(e^{\sigma^2} - 1)$
