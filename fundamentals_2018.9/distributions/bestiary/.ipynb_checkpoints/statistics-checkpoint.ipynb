# Statistics and Moments

This section contains some simplifications. I'm going to make some artificial distinctions between the terms "statistics" and "moments" and later "parameters". In practice, the lines between these concepts are often very blurry.

## Statistics

A *statistic* is any number you can calculate from your data. It can be a count, an "average", a percent, a difference, a rate, etc. So, for example, you might calculate total sales for February. That's a statistic. One conceptual stumbling block for many--and for a long time this included myself--is, "the accountant down the hall just made the same calculation. What makes my result a statistic?"

Although the specifics differ, the general answer to that question is that, your accountant is interested in reporting. They must report the sales to shareholders, they must pay taxes on the sales, or they need to inform managers of how well their sales teams did.

When doing data science, on the other hand, the goal is to use that statistic as a model to find some regularity or relationship because we want to understand the system or process sufficiently to:

1. extrapolate from the specific to the general.
2. predict the future from the past.

The first case is an example of the typical "take a random sample to understand the population" type of experimentation. We might only look at a random subset of our users because the full data is to large (or the data we have is theoretically a random sample of all possible users). The second case is when we want to look at February's purchase rate and say something about April's purchase rate. Both of these are very different from the accountant's way of looking at and using the same numbers.

And what are those numbers?

1. **Counts** - we're very often interested in counts. What was the count of customer service phone calls during an hour? What was the count of museum visitors by country? What is the number of customers last month or the number of trial accounts during the last quarter. These are often called **totals** as well in cases where count doesn't quite apply or seems weird. For example, we say "total dollars" not "the count of dollars".
2. **Averages** - Averages are just counts divided by some equalizing or normalizing factor. For example, we might take sales for last year and divide by 12 months to get "sales per month". Averages are thus closely allied with **rates** or **frequencies** ("how oftens" not radio frequencies!).
3. **Percents** or **Percentages** - We may be interested in a part/whole relationship such as the percent of site visitors who make a purchase or the percent of customers who are women. Again, these tend to bleed into the other statistics as we can talk about the percent of trial accounts that converted to paid accounts which is called a "Conversion Rate". Rates are often multipled by 100 so that that 0.02 is often expressed as a rate of 2.0. And, of course, we often do this when we write percents as "2%" or "2 percent". In different contexts, **percents** are often called **relative** frequencies

Perhaps the dirtiest secret of Data Science is that most organizations who are just starting out in data science only really need these kinds of statistics and an emailed report (and not a real time dashboard). Think of the CoNVO from the previous chapter...what is the problem--lack of knowledge--that we're trying to solve with math, statistics and programming?

## Moments

**Moments** are calculations that describe a variable's *empirical distribution*. We will have a lot to say about collecting and exploring data later but for now, just assume some data fell from the sky into your lap(top), you divided it up into ranges of values, count how many data points occur in each range and expressed those as a relative frequency. This is an approximation of the empirical distribution.

```python
%matplotlib inline
```

```python
import numpy as np
from numpy.random import seed
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
```

```python
seed( 1473727359)
```

```python
data_from_sky = norm.rvs(5.0, 5.0, 100)
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1, 1, 1)
axes.hist(data_from_sky, alpha=0.75, normed=True, color="steelblue")
axes.set_title( "Distribution of Data from Sky")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")
```

## k-th Moment

The k-th moment of the empirical distribution is:

$m_k = \frac{1}{n} \sum x_i^k$

where $n$ is the number of data points.

## First Moment 

This means that the first moment of the empirical distribution is:

$m_1 = \frac{1}{n}\sum x_i^1 = \frac{1}{n}\sum x_i$

which is also the formula for the arithmetic mean. The first moment identifies the *location* of the distribution and you'll often see software libraries and packages refer to `loc` or `location`.

```python
data1 = norm.rvs(5.0, 5.0, 100)
data2 = norm.rvs(10.0, 5.0, 100)
data3 = norm.rvs(1.0, 5.0, 100)
all_data = np.concatenate([data1, data2, data3])
values, bins = np.histogram( all_data, normed=True, bins=25)

figure = plt.figure(figsize=(20,6))
figure.suptitle( "Three Empirical Distributions with Different $m_1$")

axes = figure.add_subplot(1, 3, 1)
axes.hist(data1, bins=bins, alpha=0.50, normed=True, color="steelblue")
axes.axvline( np.mean( data1), color="black")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 2)
axes.hist(data2, bins=bins, alpha=0.50, normed=True, color="firebrick")
axes.axvline( np.mean( data2), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 3)
axes.hist(data3, bins=bins, alpha=0.50, normed=True, color="orange")
axes.axvline( np.mean( data3), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)
```

We can calculate the first moment in Python using Numpy's `mean` function:

```python
print("data1 m1:", np.mean( data1))
print("data2 m1:", np.mean( data2))
print("data3 m1:", np.mean( data3))
```

## Second Moment

The Second Moment of the empirical distribution measures how spread out the data is:

$m_2 = \frac{1}{n}\sum_i^2 = \frac{1}{n}\sum (x_i - m_1)^2$

which simplifies to the formula for *variance*. Variance is a measure of the *dispersion* of the data. Because variance is in squared units of whatever units $x$ is measured (squared feet if $x$ is in feet), we often take the square root of the variance to get the *standard deviation* which is in the same units as $x$.



```python
data1 = norm.rvs(0.0, 5.0, 100)
data2 = norm.rvs(0.0, 2.5, 100)
data3 = norm.rvs(0.0, 10.0, 100)
all_data = np.concatenate([data1, data2, data3])
values, bins = np.histogram( all_data, normed=True, bins=25)

figure = plt.figure(figsize=(20,6))
figure.suptitle( "Three Empirical Distributions with Different $m_2$")

axes = figure.add_subplot(1, 3, 1)
axes.hist(data1, bins=bins, alpha=0.50, normed=True, color="steelblue")
axes.axvline( np.mean( data1), color="black")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 2)
axes.hist(data2, bins=bins, alpha=0.50, normed=True, color="firebrick")
axes.axvline( np.mean( data2), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 3)
axes.hist(data3, bins=bins, alpha=0.50, normed=True, color="orange")
axes.axvline( np.mean( data3), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)
```

In Python we can get the second moment using Numpy's `var` function and the standard deviation using the `std` function:

```python
print("data1 m2 (std): %.4f (%.4f)" % (np.var(data1), np.std(data1)))
print("data2 m2 (std): %.4f (%.4f)" % (np.var(data2), np.std(data2)))
print("data3 m2 (std): %.4f (%.4f)" % (np.var(data3), np.std(data3)))
```

We can see that the first data set has a $m_2 = 16.00$ while the second data set has a $m_2 = 6.72$. This makes the second data set less dispersed than the second about its first moment (which is roughly the same). The third data set is more dispersed than either the first or second with $m_2 = 98.68$. As mentioned, the units of $m_2$ are in squared units. The standard deviations for each data set are given in parentheses.

Just as the first moment is often called "location", the second is often called "shape" and you can see why in the charts above. Although the three distributions have approximately the same location, their "shape"--dispersion--about the location is quite different from each other.

There are higher moments but we will not be discussing them in this book.

We will have a lot more to say about the mean, variance, etc., in later chapters. For now, we want to concentrate on the idea of moments (location and shape) and how we can use them to estimate the parameters of Mathematical Distributions and the means and variances of those distributions.

Wait, why? We just calculated the mean and variance because we calculated the first and second moments. The key point here is that we calculated the mean and variance of the *empirical distribution*. When we move to Mathematical distributions, they very often have different formulas for their means and variances.
