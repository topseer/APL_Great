# Moments

Once again our data is synthetic data for a single variable (we'll use "variable", "feature" and "factor" as well as other terms interchangeably; for now, it's "variable"). The variable is *numerical* and *continuous* (alternatives are "categorical" like Religion or Nationality; or, if the data is numerical, it may be *discrete* like counts of beans instead of continuous like inches).

As we saw in the last chapter, one way to summarize a single variable is to use a histogram. By way of review, a histogram defines intervals of values (0-10, 11-20, 21-30, etc) and counts how many values fall into each range. These counts can be expressed as *relative frequencies* (the percentage of values that fall into each range) or, what is more likely the case, a density as we described in the last section. This connects actual data with probability.

The histogram is a visual representation of the variable's values and we call it the **empirical distribution** because it comes from observed values and not, as we will see, from a model.

In this section we're going to motivate a more general way of thinking about the mean and variance of a data set and give the same formulas different names. Think of it this way: the mean and variance are the parameters of the *Gaussian* (Normal) distribution. However, it turns out the Gaussian distribution isn't the only possible distribution. There are scores of different distributions that emerge from different data generating processes. 

So we don't confuse ourselves when we're talking about the general characteristics of empirical distributions with the parameters of the Normal distribution, we call the results of the formulas something else. In other words, we don't call them the mean and the variance *until we decide we want to use the Gaussian distribution as a model for the data*.

## Calculating Moments

**Moments** are calculations that describe a variable's *empirical distribution*.  The term "moment" actually comes from physics so the analogies are often from physics as well.

The first moment of the empirical distribution is often called it's *location*. It represents the center of gravity of the data. The second moment of the empirical distribution is often called it's *scale*. It represents how spread out the data is from the center of gravity.

The k-th moment of the empirical distribution is:

$m_k = \frac{1}{n} \sum x_i^k$

where $n$ is the number of data points.

## First Moment 

The first moment of the empirical distribution is:

$m_1 = \frac{1}{n}\sum x_i^1 = \frac{1}{n}\sum x_i$

which is also the formula for the arithmetic mean. The first moment identifies the *location* of the distribution and you'll often see software libraries and packages refer to the formal function parameter as `loc` or `location`.

```python
data1 = norm.rvs(5.0, 5.0, 100)
data2 = norm.rvs(10.0, 5.0, 100)
data3 = norm.rvs(1.0, 5.0, 100)
all_data = np.concatenate([data1, data2, data3])
values, bins = np.histogram( all_data, density=True, bins=25)

figure = plt.figure(figsize=(20,6))
figure.suptitle( "Three Empirical Distributions with Different $m_1$")

axes = figure.add_subplot(1, 3, 1)
axes.hist(data1, bins=bins, alpha=0.75, density=True, color="DimGray")
axes.axvline( np.mean( data1), color="black")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 2)
axes.hist(data2, bins=bins, alpha=0.75, density=True, color="DimGray")
axes.axvline( np.mean( data2), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 3)
axes.hist(data3, bins=bins, alpha=0.75, density=True, color="DimGray")
axes.axvline( np.mean( data3), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

plt.show()
plt.close()
```

We can calculate the first moment in Python using Numpy's `mean` function:

```python
print("data1 m1:", np.mean( data1))
print("data2 m1:", np.mean( data2))
print("data3 m1:", np.mean( data3))
```

This can sometimes be a bit confusing. We're calculating the mean. What has this got to do with the first moment?

In terms of the math, absolutely nothing. This is why we use `mean` to do the calculation. In terms of semantics, however, it means that we have not decided that the normal distribution is a "good" model for our data yet and so we call it the "first moment" instead of mean.

You should get in the habit. Don't see the word "mean" and think, "oh, this is normally distributed; it has a bell shape. It should have a bell shape."

## Second Moment

The Second Moment of the empirical distribution measures how spread out the data is:

$m_2 = \frac{1}{n}\sum_i^2 = \frac{1}{n}\sum (x_i - m_1)^2$

which simplifies to the formula for *variance*. Variance is a measure of the *dispersion* of the data. Because variance is in squared units of whatever units $x$ is measured (feet squared if $x$ is in feet; pounds squared if pounds, and dollars squared if dollars), we often take the square root of the variance to get the *standard deviation* which is in the same units as $x$.

```python
data1 = norm.rvs(0.0, 5.0, 100)
data2 = norm.rvs(0.0, 2.5, 100)
data3 = norm.rvs(0.0, 10.0, 100)
all_data = np.concatenate([data1, data2, data3])
values, bins = np.histogram( all_data, density=True, bins=25)

figure = plt.figure(figsize=(20,6))
figure.suptitle( "Three Empirical Distributions with Different $m_2$")

axes = figure.add_subplot(1, 3, 1)
axes.hist(data1, bins=bins, alpha=0.75, density=True, color="DimGray")
axes.axvline( np.mean( data1), color="black")
axes.set_ylabel( "Probability Density")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 2)
axes.hist(data2, bins=bins, alpha=0.75, density=True, color="DimGray")
axes.axvline( np.mean( data2), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)

axes = figure.add_subplot(1, 3, 3)
axes.hist(data3, bins=bins, alpha=0.75, density=True, color="DimGray")
axes.axvline( np.mean( data3), color="black")
axes.set_xlabel( "X")
axes.set_xlim(-20, 20)


plt.show()
plt.close()
```

In Python we can get the second moment using Numpy's `var` function and the standard deviation using the `std` function:

```python
print("data1 m2 (std): %.4f (%.4f)" % (np.var(data1), np.std(data1)))
print("data2 m2 (std): %.4f (%.4f)" % (np.var(data2), np.std(data2)))
print("data3 m2 (std): %.4f (%.4f)" % (np.var(data3), np.std(data3)))
```

We can see that the first data set has a $m_2 = 16.00$ while the second data set has a $m_2 = 6.72$. This makes the second data set less dispersed than the second about its first moment (which is roughly the same). The third data set is more dispersed than either the first or second with $m_2 = 98.68$. As mentioned, the units of $m_2$ are in squared units. The standard deviations for each data set are given in parentheses.

Just as the first moment is often called "location", the (square root of the) second is often called "scale" (or spread) and you can see why in the charts above. Although the three distributions have approximately the same location, their "scale"--dispersion--about the location is quite different from each other.

There are higher moments but we will not be discussing them in this book.

We will have a lot more to say about the mean, variance, etc., in later chapters. For now, we want to concentrate on the idea of moments (location and scale) and how we can use them to estimate the parameters of Mathematical Distributions and the means and variances of those distributions.

Wait, what? 

It is a bit more confusing than that.

1. The mean and the variance are the parameters of the normal distribution.
2. If I have normally distributed data, I can calculate the parameters directly as the first and second moments.
3. Those moments are also the mean and variance of the data.
4. If the data are not normally distributed, I can use the first and second moments to calculate the parameters of those distributions. Additionally, I may use those parameters to calculate the mean and variance of the data (as descriptive statistics) which may not actually be the first and second moments.

It'll be a bit clearer when we look at our first distribution.

We just calculated the mean and variance because we calculated the first and second moments. The key point here is that we calculated the mean and variance of the *empirical distribution*. When we move to Mathematical distributions, they very often have different formulas for their means and variances.
